DATA TRANSFORMATION AND DATA CLEANING

```python
# removing duplicated rows
travels_data.drop_duplicates()
```

```python
def remove_spaces(text):
    if type(text) == str:
        spaces=text.strip()
        return spaces
    return text
```

```python
travels_data['Travel Start Date']=travels_data['Travel Start Date'].apply(remove_spaces)
travels_data['Travel End Date']=travels_data['Travel End Date'].apply(remove_spaces)
```

```python
travels_data.info()
```

```python
travels_data['Travel Start Date']=pd.to_datetime(travels_data['Travel Start Date'])
travels_data['Travel End Date']=pd.to_datetime(travels_data['Travel End Date'])
```

```python
travels_data.info()
```

```python
travels_data[travels_data['Purpose Of Travel'].isna()]
```

```python
travels_data['Purpose Of Travel']=travels_data['Purpose Of Travel'].fillna('Not provided')
```

```python
travels_data.info()
```

seperating states and cities

```python
def split_state(text):
    results=text.split(',')[1]
    return results

def split_city(text):
    results=text.split(',')[0]
    return results

```

```python
travels_data['state']=travels_data['Destination'].apply(split_state)
travels_data['city']=travels_data['Destination'].apply(split_city)
```

```python
travels_data
```

```python
def remove_white_spaces(name):
    result=name.replace('@','')
    result=name.replace('!','')
    return result
```

```python
travels_data['Employee']=travels_data['Employee'].apply(remove_white_spaces)
```

```python
travels_data
```

```python
def split_firstname(text):
    results=text.split(' ')[0]
    return results

def split_lastname(text):
    results=text.split(' ')[-1]
    return results
```

```python
travels_data['First Name']=travels_data['Employee'].apply(split_firstname)
travels_data['Last Name']=travels_data['Employee'].apply(split_lastname)
```

```python
travels_data
```

```python
travels_data.to_csv('cleaned_data_set.csv')
```

## US DEMOGRAPHY

```python
demography =pd.read_json("https://public.opendatasoft.com/api/explore/v2.1/catalog/datasets/us-cities-demographics/exports/json?lang=en&timezone=Africa%2FLagos")
```

```python
demography.info()
```

```python
demography.to_csv('demography.csv')
```

```python
demography['female_population']=demography['female_population'].fillna(0)
demography['male_population']=demography['male_population'].fillna(0)
demography['foreign_born']=demography['foreign_born'].fillna(0)
demography['average_household_size']=demography['average_household_size'].fillna(0)
demography['number_of_veterans']=demography['number_of_veterans'].fillna(0)
```
