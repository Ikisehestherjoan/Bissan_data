```python
#importing all the neccessary libraries  and make the neccesry connections
import pandas as pd
import psycopg2
```

```python
conn = psycopg2.connect(
                    host='localhost',
                    user='postgres',
                    password='pass',
                    database='bissan_exports'

)
cur=conn.cursor()
```

```python
query='''SELECT DISTINCT name, entitled_bonus AS stipend
            FROM employees e
            LEFT JOIN departments d
            ON e.department_id =d._id
'''
```

```python

cur.execute(query)
all_employee_stipend=cur.fetchall()
```

```python
cur.description
```

```python
employee_stipend=pd.DataFrame(all_employee_stipend, columns=['name','stipend'])
employee_stipend
```

```python
travel_dataset=pd.read_csv('cleaned_data_set.csv')

travel_dataset
```

```python
travel_dataset=travel_dataset.drop('Unnamed: 0' , axis=1) # eliminating the unawanted columns
```

```python
travel_dataset
```

```python
Employee_spending=travel_dataset[['Employee','Actual Total Expenses']]
```

```python
Employee_spending.head(5)
```

```python
Employee_spending.groupby('Employee').sum()
```

```python
employee_stipend=employee_stipend.drop_duplicates('name')
```

```python
#employee_stipend=employee_stipend.drop('index',axis=1)
employee_stipend
```

```python
Employee_spending=Employee_spending.reset_index()
```

```python
Employee_spending
```

```python
#Joining the two tables
merging_expenses=Employee_spending.merge(employee_stipend,how='inner',left_on='Employee',right_on='name')
merging_expenses
```

```python
above_stipend=merging_expenses[merging_expenses['Actual Total Expenses']>merging_expenses['stipend']]
above_stipend
#only Thomas spent above the his stipend
```
